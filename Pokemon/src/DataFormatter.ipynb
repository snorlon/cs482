This is for the express purpose of formatting LackeyCCG XML data into a csv file, excluding information we don't want

```python
import IPython.core.display
from IPython.core.display import HTML
import os;


def typeToInt(ttype):
    if(ttype.find("Colorless") != -1):
        return 0;
    if(ttype.find("Grass") != -1):
        return 1;
    if(ttype.find("Fire") != -1):
        return 2;
    if(ttype.find("Water") != -1):
        return 3;
    if(ttype.find("Lightning") != -1):
        return 4;
    if(ttype.find("Fighting") != -1):
        return 5;
    if(ttype.find("Psychic") != -1):
        return 6;
    if(ttype.find("Darkness") != -1):
        return 7;
    if(ttype.find("Metal") != -1):
        return 8;
    if(ttype.find("Dragon") != -1):
        return 9;
    if(ttype.find("Fairy") != -1):
        return 10;
    if(ttype.find("Trainer") != -1):
        return 11;
    if(ttype.find("Supporter") != -1):
        return 12;
    if(ttype.find("Stadium") != -1):
        return 12;
    if(ttype.find("Energy") != -1):
        return 14;
    
    return -1;#default to normal

formattedString = '';
cardCount = 0;

for line in open('carddata.txt', 'r'):
    #line = line.replace(", ", "_");
    line = line.replace("\t\t", "\t");
    rawData = line.split('\t')

    if rawData[0] != "Name":
        filename = rawData[2];
        typingR = rawData[3];
        typing = typeToInt(typingR);
        
        if(typing == -1):
            continue;#lets abort this iteration if we didn't get a valid typing on the card!
            
        #check if we have an image for it or not
        if os.path.isfile('images/'+filename+".jpg"):#we have an arbitrary data cap
            formattedString = formattedString + filename + ".jpg," + str(typing) + "\n";

            #Later, we can add in more elements we want to test for and format them accordingly here

            cardCount = cardCount + 1;
        
f = open('formattedData2.csv', 'w')
f.write(formattedString);
f.close();

print formattedString;

#HTML(readable)

print cardCount;#this is how many cards we have for our data set
```
